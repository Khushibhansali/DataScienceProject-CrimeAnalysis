```python
'''
Project 4: data patterns and maps
'''

#Part 1
#!pip install folium
import folium
import requests
import pandas
from ipykernel import kernelapp as app

arrest_table = pandas.read_csv("https://cmsc320.github.io/files/BPD_Arrests.csv")

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)
arrest_table.head()
```

```python
#Part 2
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)
map_osm
```

```python
"""
PART 3: short prose

The arrest table dataframe contains 63892 entries. The dataframe records gender, location, neighborhood,
crime offense, and much more. Of this dataframe, I chose to look at the first thousand entries to see
how many crimes of narcotics, common assault, recovered property, and other occurred per gender. To depict this,
the map shows

    light gray:     79-Other Male
    black:          79-Other Female
    light green:    87-Narcotics Male
    dark green:     87-Narcotics Female
    blue:           4E-Common Assault Male
    darkblue:       4E-Common Assault Female
    purple:         81-Recovered Property Male
    pink:           81-Recovered Property Female

The purpose of selecting these fields ('incidentOffense','lat', 'long', 'sex'), was to see that how
many of these crimes occurred in maryland, virginia, and dc by gender. Additionally, by restricting
the sample to the first 1000, I wanted to see how representative the sample would be of the population.

After seeing the statistics and map, the light green and blue color cover the majority of the map.
These represent 87-Narcotics Male and 4E-Common Assault Male respectively. This proves that the
sample is indeed representative of the population since the population also shows that the majority
of offenses are commited are 87-Narcotics and 4E-Common Assault.

"""
```

```python
#Part 3 - group dataframe by incident offense, location and gender of offender

loc = arrest_table[['incidentOffense','lat', 'long', 'sex']]
other = 0
narcotics = 0
assault = 0
recovered = 0

#In a sample of 1000 people, this is the percentage of each crime occurence
for i in range(0, 1000):
     if (loc.iloc[i]['incidentOffense'] == "79-Other"):
        other = other + 1
     elif (loc.iloc[i]['incidentOffense'] == "87-Narcotics"):
        narcotics = narcotics + 1
     elif (loc.iloc[i]['incidentOffense'] == "4E-Common Assault"):
        assault = assault + 1
     elif (loc.iloc[i]['incidentOffense'] == "81-Recovered Property"):
        recovered = recovered + 1

other = other/10
narcotics = narcotics/10
assault = assault / 10
recovered = recovered/10

print("In a sample of first 1000 people,")
print("The offense named Other makes up " + str(other) + "% of crime")
print("Narcotics makes up " + str(narcotics) + "% of crime")
print("Assault makes up " + str(assault) + "% of crime")
print("Recovered Property makes up " + str(recovered) + "% of crime")

#In the entire population, this is the percentage of each crime occurence
other = 0
narcotics = 0
assault = 0
recovered = 0

for i in range(0, len(loc)):
     if (loc.iloc[i]['incidentOffense'] == "79-Other"):
        other = other + 1
     elif (loc.iloc[i]['incidentOffense'] == "87-Narcotics"):
        narcotics = narcotics + 1
     elif (loc.iloc[i]['incidentOffense'] == "4E-Common Assault"):
        assault = assault + 1
     elif (loc.iloc[i]['incidentOffense'] == "81-Recovered Property"):
        recovered = recovered + 1

other = other*100/len(loc)
narcotics = narcotics*100/len(loc)
assault = assault*100 / len(loc)
recovered = recovered*100/len(loc)

print()
print("----------------------------------------------")
print()

print("In terms of the whole population,")
print("The offense named Other makes up " + str(other) + "% of crime")
print("Narcotics makes up " + str(narcotics) + "% of crime")
print("Assault makes up " + str(assault) + "% of crime")
print("Recovered Property makes up " + str(recovered) + "% of crime")
```

```python
#colored offense differently based on gender
for i in range(0, 1000):

    if (loc.iloc[i]['incidentOffense'] == "79-Other" and loc.iloc[i]['sex'] == "M"):
         folium.Marker(
              location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
              popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
              icon=folium.Icon(color="lightgray"),
            ).add_to(map_osm)

    elif (loc.iloc[i]['incidentOffense'] == "79-Other" and loc.iloc[i]['sex'] == "F"):
            folium.Marker(
              location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
              popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
              icon=folium.Icon(color="black"),
            ).add_to(map_osm)

    elif(loc.iloc[i]['incidentOffense'] == "87-Narcotics" and loc.iloc[i]['sex'] == "M"):
        folium.Marker(
          location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
          popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
          icon=folium.Icon(color="lightgreen"),
        ).add_to(map_osm)

    elif(loc.iloc[i]['incidentOffense'] == "87-Narcotics" and loc.iloc[i]['sex'] == "F"):
        folium.Marker(
          location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
          popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
          icon=folium.Icon(color="darkgreen"),
        ).add_to(map_osm)

    elif(loc.iloc[i]['incidentOffense'] == "4E-Common Assault" and loc.iloc[i]['sex'] == "M"):
        folium.Marker(
          location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
          popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
          icon=folium.Icon(color="blue"),
        ).add_to(map_osm)

    elif(loc.iloc[i]['incidentOffense'] == "4E-Common Assault" and loc.iloc[i]['sex'] == "F"):
        folium.Marker(
          location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
          popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
          icon=folium.Icon(color="darkblue"),
        ).add_to(map_osm)

    elif(loc.iloc[i]['incidentOffense'] == "81-Recovered Property" and loc.iloc[i]['sex'] == "M"):
        folium.Marker(
          location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
          popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
          icon=folium.Icon(color="purple"),
        ).add_to(map_osm)

    elif(loc.iloc[i]['incidentOffense'] == "81-Recovered Property" and loc.iloc[i]['sex'] == "F"):
        folium.Marker(
          location=[loc.iloc[i]['lat'], loc.iloc[i]['long']],
          popup="Offense: " + loc.iloc[i]['incidentOffense'] + " Gender: " + loc.iloc[i]['sex'],
          icon=folium.Icon(color="pink"),
        ).add_to(map_osm)

# Show the map again
map_osm
```
